```python
from datetime import date
import random
import pandas as pd
```

```python
def pick_date(): 
    start_date = date.today().replace(day=1, month=1).toordinal()
    end_date = date.today().replace(day=31, month=12).toordinal()
    random_day = date.fromordinal(random.randint(start_date, end_date))
    return random_day
```

```python
patient_ids = []
er_visits = [] 
hospitalizations = []

num_patients = 100

for i in range(num_patients):
    v = random.randint(0,20)
    h = random.randint(0,v)
    
    patient_ids.append(i)
    er_visits.append(v)
    hospitalizations.append(h)
        
```

```python
df = pd.DataFrame({"patient_id": patient_ids, "er_visits": er_visits, "hospitalizations": hospitalizations})
```
