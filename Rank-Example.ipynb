```python
from datetime import date
import random
import pandas as pd
```

```python
def pick_date(): 
    start_date = date.today().replace(day=1, month=1).toordinal()
    end_date = date.today().replace(day=31, month=12).toordinal()
    random_day = date.fromordinal(random.randint(start_date, end_date))
    return random_day
```

```python
patient_ids = []
er_visits = [] 
hospitalizations = []

num_patients = 100

for i in range(num_patients):
    v = random.randint(0,100)
    h = random.randint(0,v)
    
    patient_ids.append(i)
    er_visits.append(v)
    hospitalizations.append(h)
        
```

```python
df = pd.DataFrame({"patient_id": patient_ids, "er_visits": er_visits, "hospitalizations": hospitalizations})
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("""
SELECT 
    patient_id, 
    er_visits,
    hospitalizations,
    DENSE_RANK() OVER (ORDER BY er_visits DESC) AS visits_rnk,
    DENSE_RANK() OVER (ORDER BY hospitalizations DESC) AS hsp_rnk
FROM df
    
""")
```

```python
visit_dates = []
hospitalized = []
event_ids = []

num_dates = 100

for i in range(num_dates):
    num_events = random.randint(1,100)
    for j in range(num_events):
        event_ids.append(i)
        visit_dates.append(pick_date())
        hospitalized.append(random.choice([1, None]))
```

```python
df = pd.DataFrame({"event_id": event_ids, "visit_date": visit_dates, "hospitalized": hospitalized})
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("""
SELECT
    visit_date, 
    COUNT(visit_date), 
    COUNT(hospitalized) as num_hospitalizations
FROM
    df
GROUP BY 
    visit_date
ORDER BY 
    num_hospitalizations DESC
""")
```
