#### Illustration of CROSS JOIN (or JOIN with no ON condition)

```python
import pandas as pd
```

```python
suits = pd.DataFrame({'Ordinal':[4, 3, 2, 1], 'S':['C', 'D', 'H', 'S']})
cards = pd.DataFrame({'Ordinal':[4, 3, 2, 1], 'F':['J', 'Q', 'K', 'A']})
```

```python
suits
```

```python
cards
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("""
SELECT 
    *
FROM
    suits s
JOIN
    cards c
ON 
    s.Ordinal = c.Ordinal
""")
```

```python
pysqldf("""
SELECT 
     c.F, s.S
FROM
    suits s
JOIN
    cards c
""")
```

```python
pysqldf("""
SELECT 
    c.F, s.S
FROM
    suits s
CROSS JOIN
    cards c
""")
```

```python
# self-joins, using a cross join
# what if I wanted to list each suit, and all suits that are below it in Ordinal rank?
```

```python
pysqldf("""
SELECT 
     s1.S as S1, s2.S as S2
FROM
    suits s1
JOIN
    suits s2
ON 
    s1.Ordinal < s2.Ordinal
""")
```
